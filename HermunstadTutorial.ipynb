```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.special import expit
```

```python
frequencies=np.arange(0,1000)
```

```python
field_pdf = stats.norm(400, 100)
forest_pdf=stats.norm(600,100)

plt.plot(frequencies, field_pdf.pdf(frequencies),color='lightcoral')
plt.plot(frequencies, forest_pdf.pdf(frequencies),color='darkorchid')
plt.fill_between(frequencies,field_pdf.pdf(frequencies),color='lightcoral',alpha=0.5)
plt.fill_between(frequencies, forest_pdf.pdf(frequencies),color='darkorchid',alpha=0.5)
plt.title('Frequencies of flowers in the field and forest')
plt.xlabel('Frequency')
```

```python
def inverse_sigmoid(x, a, b, c, d):
    """ General sigmoid function
    a adjusts amplitude
    b adjusts y offset
    c adjusts x offset
    d adjusts slope """
    y = ((a-b) / (1 + np.exp(x-(c/2))**d)) + b
    return y

def sigmoid(x, a, b, c, d):
    """ General sigmoid function
    a adjusts amplitude
    b adjusts y offset
    c adjusts x offset
    d adjusts slope """
    #y = ((a-b) / (1 + np.exp(-(x+(c/2))**d))) + b
    y=(a-b)/(1+np.exp(-x-(c))**d)+b
    return y



field_sigmoid=inverse_sigmoid(frequencies,100,10,900,0.01)
forest_sigmoid=sigmoid(frequencies,100,10,-150,0.01)

plt.plot(frequencies,field_sigmoid)
plt.plot(frequencies,forest_sigmoid)
```
